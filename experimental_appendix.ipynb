## 1. Data Imputation: Global Median vs. Age-Group Median

* **Hypothesis:** Missing mileage values can be treated as a missing-at-random (MAR) variable and filled using the global dataset median[cite: 1].
* **Observation:** Global imputation introduced significant bias. Newer vehicles were assigned unnaturally high mileage, while vintage vehicles were assigned mileage far below their likely utility[cite: 1].
* **Conclusion/Pivot:** Mileage is highly dependent on vehicle age[cite: 1]. We pivoted to **Median Imputation by Age-Group** to preserve the natural correlation between time and usage[cite: 1, 2].

## 2. Outlier Detection: Z-Score vs. Domain-Based Filtering

* **Hypothesis:** Using a statistical Z-score threshold (Z > 3) will effectively remove data entry noise from the `price` column[cite: 1].
* **Observation:** The price distribution is heavily right-skewed[cite: 1]. Z-score filtering automatically removed high-value luxury cars and "supercars"[cite: 1]. These are not errors; they are valid, high-impact market entries.
* **Conclusion/Pivot:** Switched to a **Manual Domain Filter** ($500 < Price < $200,000) based on market research[cite: 1]. This allowed us to keep the "market tail" while still removing obvious data entry errors like $0 prices[cite: 1, 2].

## 3. Visualization: Linear vs. Log-Linear Scaling

* **Hypothesis:** A standard linear scatter plot will reveal the relationship between `max_power` and `price`[cite: 1].
* **Observation:** Due to the right-skewness of price data[cite: 1], the linear plot resulted in a dense, unreadable cluster at the bottom-left of the chart, obscuring the trend.
* **Conclusion/Pivot:** Applied a **Log Transformation** to the price variable[cite: 1]. This corrected for skewness and revealed a clearer, linear depreciation trend as suggested by industry standards[cite: 1, 2].

## 4. Feature Priority: Mileage vs. Vehicle Age

* **Hypothesis:** Mileage is the primary negative driver of price in the used car market[cite: 1].
* **Observation:** While mileage correlates negatively, it proved to be a weaker and noisier predictor than **Vehicle Age**[cite: 1].
* **Analysis:** Mileage data in the raw set showed signs of reporting inconsistency and noise, whereas the manufacturing year (Age) was a more reliable integrity point[cite: 1].
* **Final Decision:** Prioritized "Vehicle Age" as the second most important feature after "Max Power," aligning our findings with the literature review of **Xinru Chen et al.**[cite: 1, 2].
